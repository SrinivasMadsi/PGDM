```python
# Function to extract primary, secondary and tertiary sectors from all datasets
def sectors_data(data):    
    new_df = data.loc[(data['Item']=='Primary')|(data['Item']=='Secondary')|(data['Item']=='Tertiary')][['Item','state','2014-15']] 
    return new_df
```

```python
data = [andhra_pradesh,arunachal_pradesh,assam,bihar,chhattisgarh,goa,gujarat,haryana,himachal_pradesh,jharkhand,karnataka,kerala,
        madya_pradesh,maharashtra,manipur,meghalaya,mizoram,nagaland,odisha,punjab,rajasthan,sikkim,tamilnadu,telangana,
        tripura,uttar_pradesh,uttarakhand]
sectors = pd.DataFrame(columns=['Item','state','2014-15'])
for x in data:
    sectors = pd.concat([sectors,sectors_data(x)],axis=0)
    
```

```python
sectors
```

```python
# Finding the total of Primary+Secondary+Tertiary for all the states
state_total = pd.DataFrame(sectors.groupby('state')['2014-15'].sum().reset_index())
state_total= state_total.rename(columns={'2014-15':'total_value'})
```

```python
# Finding the percentage of each sector for individual state
sector_data1 = pd.merge(sectors,state_total,on='state',how='left')
sector_data1['Sectors_Percent'] = (sector_data1['2014-15']*100)/(sector_data1['total_value'])
sector_data2 = sector_data1[['Item','state','Sectors_Percent']]
sector_data2
```
